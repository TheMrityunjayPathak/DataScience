<h2 style="color:#6a66bd; text-align:left;">Python Comments</h2>

<p style="color:#6a66bd;">
<b>What is Comments?</b>
<br>
• Comments can be used to explain Python code.
<br>
• Comments can be used to make the code more readable.
<br>
• Comments can be used to prevent execution when testing code.
<br>
<br>
<b>Creating a Comment</b>
<br>
• Comments starts with # and Python will ignore them.
<code>
#This is a comment
</code>
• Comments can be placed at the end of a line and Python will ignore the rest of the line.
<code>
print("Hello, World!") #This is a comment
</code>
• A comment does not have to be text that explains the code, it can also be used to prevent Python from executing code.
<code>
#print("Hello, World!")
</code>
<br>
<b>Multiline Comments</b>
<br>
• Python does not really have a syntax for multiline comments.
<br>
• To add a multiline comment you could insert a # for each line.
<code>
#This is a comment
#written in
#more than just one line
</code>
• You can also use a multiline string (docstring).
<code>
"""
This is a comment
written in
more than just one line
"""
</code>
</p>

<h2 style="color:#6a66bd; text-align:left;">Python Variables</h2>

<p style="color:#6a66bd;">
<b>What is Variables?</b>
<br>
• Variables are containers for storing data values.
<br>
<br>
<b>Creating a Variable</b>
<br>
• Python has no command for declaring a variable.
<br>
• A variable is created the moment you first assign a value to it.
<code>
x = 5
y = "John"
print(x)
print(y)
</code>
• Variables do not need to be declared with any particular type and can even change type after they have been set.
<code>
x = 4       #x is of type int
x = "Sally" #x is now of type str
print(x)
</code>
<br>
<b>Casting</b>
<br>
• If you want to specify the data type of a variable, this can be done with casting.
<code>
x = str(3)    #x will be '3'
y = int(3)    #y will be 3
z = float(3)  #z will be 3.0
</code>
<br>
<b>Get the Type</b>
<br>
• You can get the data type of a variable with the type() function.
<code>
x = 5
y = "John"
print(type(x))
print(type(y))
</code>
<br>
<b>Case Sensitive</b>
<br>
• Variable names are case-sensitive.
<code>
a = 4
A = "Sally"
#A will not overwrite a
</code>
<br>
<b>Naming Python Variable</b>
<br>
• A variable name must start with a letter or the underscore character.
<br>
• A variable name cannot start with a number.
<br>
• A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ ).
<br>
• Variable names are case-sensitive (age, Age and AGE are three different variables).
<br>
• A variable name cannot be any of the Python keywords.
<br>
<br>
<b>Legal Variable Names</b>
</p>

<div style="background-color:seagreen; border-radius:4px;"><p style="color:white; padding:10px;">
myvar = "John"
<br>
my_var = "John"
<br>
_my_var = "John"
<br>
myVar = "John"
<br>

<br>
myvar2 = "John"
</p></div>

<p style="color:#6a66bd;">
<b>ilLegal Variable Names</b>

<br>

<br>

<p style="color:#6a66bd;">
<b>Multi Words Variable Names</b>
<br>
• Variable names with more than one word can be difficult to read.
<br>
• There are several techniques you can use to make them more readable.
<br>
<b>Camel Case</b>
<br>
• Each word except the first starts with a capital letter.
<br>
<code>myVariableName = "John"</code>
<br>
<b>Pascal Case</b>
<br>
• Each word starts with a capital letter.
<br>
<code>MyVariableName = "John"</code>
<br>
<b>Snake Case</b>
<br>
• Each word is separated by an underscore character.
<br>
<code>my_variable_name = "John"</code>
<br>
<br>
<b>Assign Multiple Values</b>
<br>
• Python allows you to assign values to multiple variables in one line.
<code>
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
</code>
<b>Note :</b> Make sure the number of variables matches the number of values or else you will get an error.
<br>
• You can assign the same value to multiple variables in one line also.
<code>
x = y = z = "Orange"
print(x)
print(y)
print(z)
</code>
<br>
<b>Output Variables</b>
<br>
• The Python print() function is often used to output variables.
<code>
x = "Python is awesome"
print(x)
</code>
• In the print() function, you output multiple variables separated by a comma.
<code>
x = "Python"
y = "is"
z = "awesome"
print(x, y, z)
</code>
• You can also use the + operator to output multiple variables.
<code>
x = "Python "
y = "is "
z = "awesome"
print(x + y + z)
</code>
• For numbers, the + character works as a mathematical operator.
<code>
x = 5
y = 10
print(x + y)
</code>
• The best way to output multiple variables in the print() function is to separate them with commas which even support different data types.
<code>
x = 5
y = "John"
print(x, y)
</code>
<br>
<b>Global Variables</b>
<br>
• Variables that are created outside of a function are known as global variables.
<br>
• Global variables can be used everywhere both inside of function and outside.
<code>
x = "awesome"
def myfunc():
  print("Python is " + x)
myfunc()
</code>
• If you create a variable with the same name inside a function, this variable will be local and can only be used inside the function.
<br>
• The global variable with the same name will remain as it was and with the original value.
<code>
x = "awesome"
def myfunc():
  x = "fantastic"
  print("Python is " + x)
myfunc()
print("Python is " + x)
</code>
• Normally, when you create a variable inside a function that variable is local and can only be used inside that function.
<br>
• To create a global variable inside a function you can use the global keyword.
<code>
def myfunc():
  global x
  x = "fantastic"
myfunc()
print("Python is " + x)
</code>
• Also, use the global keyword if you want to change a global variable inside a function.
<code>
x = "awesome"
def myfunc():
  global x
  x = "fantastic"
myfunc()
print("Python is " + x)
</code>
</p>

<h2 style="color:#6a66bd; text-align:left;">Python Casting</h2>

<p style="color:#6a66bd;">
<b>Specify a Variable Type</b>
<br>
• There may be times when you want to specify a type on to a variable. This can be done with casting.
<br>
• Python is an object-orientated language and as such it uses classes to define data types including its primitive types.
<br>
• Casting in python is therefore done using constructor functions:
<br>
<b>int() -</b> Constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number).
<code>
x = int(1)   #x will be 1
y = int(2.8) #y will be 2
z = int("3") #z will be 3
</code>
<b>float() -</b> Constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer).
<code>
x = float(1)     #x will be 1.0
y = float(2.8)   #y will be 2.8
z = float("3")   #z will be 3.0
w = float("4.2") #w will be 4.2
</code>
<b>str() -</b> Constructs a string from a wide variety of data types, including strings, integer literals and float literals.
<code>
x = str("s1") #x will be 's1'
y = str(2)    #y will be '2'
z = str(3.0)  #z will be '3.0'
</code>

<h2 style="color:#6a66bd; text-align:left;">Python Strings</h2>

<p style="color:#6a66bd;">
<b>Strings</b>
<br>
• Strings in python are surrounded by either single quotation marks or double quotation marks.
<br>
• 'hello' is the same as "hello".
<br>
• You can display a string literal with the print() function.
<code>
print("Hello")
print('Hello')
</code>
<br>
<b>Quotes Inside Quotes</b>
<br>
• You can use quotes inside a string, as long as they don't match the quotes surrounding the string.
<code>
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')
</code>
<br>
<b>Assign String to a Variable</b>
<br>
• Assigning a string to a variable is done with the variable name followed by an equal sign and the string.
<code>
a = "Hello"
print(a)
</code>
<br>
<b>Multiline Strings</b>
<br>
• You can assign a multiline string to a variable by using three quotes.
<code>
#You can use three double quotes
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
</code>
<code>
#Or three single quotes
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
print(a)
</code>
<b>Note :</b> In the result the line breaks are inserted at the same position as in the code.
<br>
<br>
<b>Strings are Arrays</b>
<br>
• Like many other popular programming languages strings in Python are arrays of bytes representing unicode characters.
<br>
• However Python does not have a character data type a single character is simply a string with a length of 1.
<br>
• Square brackets can be used to access elements of the string.
<code>
a = "Hello, World!"
print(a[1])
</code>
<b>Note :</b> Remember that the first character has the position 0.
    

</p>

<p style="color:#6a66bd;">
<b>Looping Through a String</b>
<br>
• Since strings are arrays, we can loop through the characters in a string with a for loop.
<code>
for x in "banana":
  print(x)
</code>
<br>
<b>String Length</b>
<br>
• To get the length of a string use the len() function.
<code>
a = "Hello, World!"
print(len(a))
</code>
<br>
<b>Check String</b>
<br>
• To check if a certain phrase or character is present in a string, we can use the keyword <code>in</code>.
<code>
txt = "The best things in life are free!"
print("free" in txt)
</code>
• To check if a certain phrase or character is NOT present in a string, we can use the keyword <code>not in</code>.
<code>
txt = "The best things in life are free!"
print("expensive" not in txt)
</code>
<br>
<b>String Slicing</b>
<br>
• You can return a range of characters by using the slice syntax.
<br>
• Specify the start index and the end index separated by a colon to return a part of the string.
<code>
b = "Hello, World!"
print(b[2:5])
</code>
<b>Note :</b> The first character has index 0.
<br>
<b>Slice From the Start</b>
<br>
• By leaving out the start index the range will start at the first character.
<code>
b = "Hello, World!"
print(b[:5])
</code>
<b>Slice To the End</b>
<br>
• By leaving out the end index the range will go to the end.
<code>
b = "Hello, World!"
print(b[2:])
</code>
<b>Negative Indexing</b>
<br>
• Use negative indexes to start the slice from the end of the string.
<code>
b = "Hello, World!"
print(b[-5:-2])
</code>
<br>
<b>String Concatenation</b>
<br>
• To concatenate or combine two strings you can use the + operator.
<br>
• Merge variable a with variable b into variable c.
<code>
a = "Hello"
b = "World"
c = a + b
print(c)
</code>
• To add a space between them add a " ".
<code>
a = "Hello"
b = "World"
c = a + " " + b
print(c)
</code>
<br>
<b>String Format</b>
<br>
• F-String was introduced in Python 3.6 and is now the preferred way of formatting strings.
<br>
• To specify a string as an f-string simply put an f in front of the string literal and add curly brackets {} as placeholders for variables and other operations.
<code>
age = 36
txt = f"My name is John, I am {age}"
print(txt)
</code>
<b>Placeholders and Modifiers</b>
<br>
• A placeholder can contain variables, operations, functions and modifiers to format the value.
<code>
price = 59
txt = f"The price is {price} dollars"
print(txt)
</code>
<b>Note :</b> Price is the placeholder here.
<br>
• A placeholder can include a modifier to format the value.
<br>
• A modifier is included by adding a colon : followed by a legal formatting type like .2f which means fixed point number with 2 decimals.
<code>
price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)
</code>
• A placeholder can contain Python code like math operations.
<code>
txt = f"The price is {20 * 59} dollars"
print(txt)
</code>

<br>

<h2 style="color:#6a66bd; text-align:left;">Python Print</h2>

<p style="color:#6a66bd;">
• The print() function prints the specified message to the screen.
<br>
• The message can be a string or any other object the object will be converted into a string before written to the screen.

<b>Parameter Values</b>
<br>

<br>

<br>
• end (Optional) - Specify what to print at the end. Default is '\n' (new line).
<code>

</code>

<b>Arithmetic Operation in Print</b>
<br>
• Print follows PEMDAS Rule for performing arithmetic operation :
<br>
P = Parenthesis
<br>
E = Exponent
<br>
M = Multiply
<br>
D = Division
<br>
A = Addition
<br>
S = Subtraction
<br>
<br>
<b>Booleans in Print</b>
<br>
• When you add booleans adding False is equivalent to adding 0 and adding True is equivalent to adding 1.
<code>

print(False + True) #Output is 1
print(True + True) #Output is 2

</code>

<code>

</code>
• If you multiply a string by a boolean with value True, it just returns that same string.
<code>
print("abc" * True) #Output is abc
</code>
• If you multiply a string by a boolean with value False, it returns an empty string with length 0.
<code>
print(len("abc" * False)) #Output is 0
</code>

<h2 style="color:#6a66bd; text-align:left;">Python List</h2>

<p style="color:#6a66bd;">
• Lists are used to store multiple items in a single variable.
<br>
• Lists are created using square brackets.
<code>
thislist = ["apple", "banana", "cherry"]
print(thislist)
</code>
<br>
<b>List Items</b>
<br>
• List items are ordered, changeable and allow duplicate values.
<br>
• List items are indexed, the first item has index [0], the second item has index [1] etc.
<br>
<br>
<b>Ordered</b>
<br>
• When we say that lists are ordered, it means that the items have a defined order and that order will not change.
<br>
• If you add new items to a list, the new items will be placed at the end of the list.
<br>
<br>
<b>Changeable</b>
<br>
• The list is changeable meaning that we can change, add and remove items in a list after it has been created.
<br>
<br>
<b>Allow Duplicates</b>
<br>
• Since lists are indexed lists can have items with the same value.
<code>
thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
</code>
<br>
<b>List Length</b>
<br>
• To determine how many items a list has use the len() function.
<code>
thislist = ["apple", "banana", "cherry"]
print(len(thislist))
</code>
<br>
<b>List Items - Data Types</b>
<br>
• List items can be of any data type.
<code>
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
</code>
• A list can contain different data types.
<code>
list1 = ["abc", 34, True, 40, "male"]
</code>
<br>
<b>list() Constructor</b>
<br>
• It is also possible to use the list() constructor when creating a new list.
<code>
thislist = list(("apple", "banana", "cherry")) #note the double round-brackets
print(thislist)
</code>
<br>
<b>Access List Items</b>
<br>
• List items are indexed and you can access them by referring to the index number.
<code>
thislist = ["apple", "banana", "cherry"]
print(thislist[1])
</code>
<b>Note :</b> The first item has index 0.
</p>



<p style="color:#6a66bd;">
<b>Negative Indexing</b>
<br>
• Negative indexing means start from the end.
<br>
• -1 refers to the last item, -2 refers to the second last item etc.
<code>
thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
</code>
<br>
<b>Range of Indexes</b>
<br>
• You can specify a range of indexes by specifying where to start and where to end the range.
<br>
• When specifying a range the return value will be a new list with the specified items.
<code>
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
</code>
<b>Note :</b> The search will start at index 2 (included) and end at index 5 (not included).
<br>
• By leaving out the start value the range will start at the first item.
<code>
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])
</code>
• By leaving out the end value the range will go on to the end of the list.
<code>
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
</code>
<br>
<b>Range of Negative Indexes</b>
<br>                                
• Specify negative indexes if you want to start the search from the end of the list.
<code>
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
</code>
<br>
<b>Check if Item Exists</b>
<br>
• To determine if a specified item is present in a list use the in keyword.
<code>
thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
  print("Yes, 'apple' is in the fruits list")
</code>
<br>
<b>Change List Items</b>
<br>
• To change the value of a specific item refer to the index number.
<code>
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
</code>
<br>
<b>Change a Range of Item Values</b>
<br>
• To change the value of items within a specific range define a list with the new values and refer to the range of index numbers where you want to insert the new values.
<code>
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)
</code>
• If you insert more items than you replace the new items will be inserted where you specified and the remaining items will move accordingly.
<code>
thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant", "watermelon"]
print(thislist)
</code>
<b>Note :</b> The length of the list will change when the number of items inserted does not match the number of items replaced.
<br>
• If you insert less items than you replace the new items will be inserted where you specified and the remaining items will move accordingly.
<code>
thislist = ["apple", "banana", "cherry"]
thislist[1:3] = ["watermelon"]
print(thislist)
</code>
<br>
<b>Insert Items</b>
<br>
• To insert a new list item without replacing any of the existing values, we can use the insert() method.
<br>
• The insert() method inserts an item at the specified index.
<code>

thislist.insert(2, "watermelon")
print(thislist)
</code>

<b>Append Items</b>

thislist = ["apple", "banana", "cherry"]

<b>Extend List</b>

• To append elements from another list to the current list use the extend() method.

thislist = ["apple", "banana", "cherry"]

<br>
<b>Add Any Iterable</b>
<br>
• The extend() method does not have to append lists, you can add any iterable object (tuples, sets, dictionaries etc.).

thislist = ["apple", "banana", "cherry"]
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(thislist)

<b>Counting List Items</b>

fruits = ['apple', 'banana', 'cherry']

</code>
<br>
<b>Indexing List Items</b>

fruits = [4, 55, 64, 32, 16, 32]

print(x)
</code>
<b>Note :</b> The index() method only returns the first occurrence of the value.
<br> 
<br>
<b>Reversing the List</b>

fruits = ['apple', 'banana', 'cherry']
fruits.reverse()
fruits
</code>
<br>
<b>Remove Specified Item</b>
<br>
• The remove() method removes the specified item.

thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)
</code>
• If there are more than one item with the specified value the remove() method removes the first occurrence.

thislist = ["apple", "banana", "cherry", "banana", "kiwi"]

print(thislist)
</code>

<b>Remove Specified Index</b>
<br>
• The pop() method removes the specified index.

thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(thislist)
</code>

thislist = ["apple", "banana", "cherry"]
thislist.pop()
print(thislist)
</code>
• The del keyword also removes the specified index.

thislist = ["apple", "banana", "cherry"]
del thislist[0]
print(thislist)
</code>
• The del keyword can also delete the list completely.
<code>
thislist = ["apple", "banana", "cherry"]
del thislist
</code>
<br>
<b>Clear the List</b>
<br>
• The clear() method empties the list.
<br>
• The list still remains but it has no content.
<code>
thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(thislist)
</code>
<br>
<b>Loop Through a List</b>
<br>
• You can loop through the list items by using a for loop.
<code>
thislist = ["apple", "banana", "cherry"]
for x in thislist:
  print(x)
</code>
<br>
<b>Loop Through the Index Numbers</b>
<br>
• You can also loop through the list items by referring to their index number.
<br>
• Use the range() and len() functions to create a suitable iterable.
<code>
thislist = ["apple", "banana", "cherry"]
for i in range(len(thislist)):
  print(thislist[i])
</code>
<br>
<b>List Comprehension</b>
<br>
• List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.
<code>
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
</code>
<b>Syntax</b>
<br>
<code>list = [expression for item in iterable if condition == True]</code>
<br>
<b>Note :</b> The return value is a new list leaving the old list unchanged.
<br>
<b>Condition</b>
<br>
• The condition is like a filter that only accepts the items that valuate to True.
<code>
newlist = [x for x in fruits if x != "apple"]
</code>
<b>Note :</b> The condition is optional and can be omitted.
<br>
<b>Iterable</b>
<br>
• The iterable can be any iterable object like a list, tuple, set, range etc.
<code>
newlist = [x for x in range(10)]
</code>
<br>
<b>Sort Lists</b>

</code>

<b>Copy Lists</b>
<br>

<br>
<b>Join Lists</b>

• There are several ways to join or concatenate two or more lists in Python.

<h2 style="color:#6a66bd; text-align:left;">Python Tuple</h2>

<p style="color:#6a66bd;">
• Tuples are used to store multiple items in a single variable.
<br>
• A tuple is a collection which is ordered and unchangeable.
<br>
• Tuples are written with round brackets.
<code>
thistuple = ("apple", "banana", "cherry")
print(thistuple)
</code>
<br>
<b>Tuple Items</b>
<br>
• Tuple items are ordered, unchangeable and allow duplicate values.
<br>
• Tuple items are indexed, the first item has index [0], the second item has index [1] etc.
<br>
<br>
<b>Ordered</b>
<br>
• When we say that tuples are ordered it means that the items have a defined order and that order will not change.
<br>
<br>
<b>Unchangeable</b>
<br>
• Tuples are unchangeable meaning that we cannot change, add or remove items after the tuple has been created.
<br>
<br>
<b>Allow Duplicates</b>
<br>
• Since tuples are indexed they can have items with the same value.
<code>
thistuple = ("apple", "banana", "cherry", "apple", "cherry")
print(thistuple)
</code>
<br>
<b>Tuple Length</b>
<br>
• To determine how many items a tuple has use the len() function.
<code>
thistuple = ("apple", "banana", "cherry")
print(len(thistuple))
</code>
<br>
<b>Create Tuple With One Item</b>
<br>
• To create a tuple with only one item, you have to add a comma after the item otherwise Python will not recognize it as a tuple.
</p>

<div style="background-color:seagreen; border-radius:4px;"><p style="color:white; padding:10px;">
thistuple = ("apple",)
<br>
print(type(thistuple))
</p></div>

<p style="color:#6a66bd;">
• If the comma is not added, Python will give an error.
</p>

<div style="background-color:crimson; border-radius:4px;"><p style="color:white; padding:10px;">
thistuple = ("apple")
<br>
print(type(thistuple))
</p></div>

<p style="color:#6a66bd;">
<b>Tuple Items - Data Types</b>
<br>
• Tuple items can be of any data type.
<code>
tuple1 = ("apple", "banana", "cherry")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)
</code>
• A tuple can contain different data types.
<code>
tuple1 = ("abc", 34, True, 40, "male")
</code>
<br>
<b>tuple() Constructor</b>
<br>
• It is also possible to use the tuple() constructor to make a tuple.
<code>
thistuple = tuple(("apple", "banana", "cherry")) #note the double round-brackets
print(thistuple)
</code>
<br>
<b>Access Tuple Items</b>
<br>
• You can access tuple items by referring to the index number inside square brackets.
<code>
thistuple = ("apple", "banana", "cherry")
print(thistuple[1])
</code>
<b>Note :</b> The first item has index 0.
<br>
<br>
<b>Negative Indexing</b>
<br>
• Negative indexing means start from the end.
<br>
• -1 refers to the last item, -2 refers to the second last item etc.
<code>
thistuple = ("apple", "banana", "cherry")
print(thistuple[-1])
</code>
<br>
<b>Range of Indexes</b>
<br>
• You can specify a range of indexes by specifying where to start and where to end the range.
<br>
• When specifying a range, the return value will be a new tuple with the specified items.
<code>
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])
</code>
<b>Note :</b> The search will start at index 2 (included) and end at index 5 (not included).
<br>
• By leaving out the start value, the range will start at the first item.
<code>
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[:4])
</code>
• By leaving out the end value, the range will go on to the end of the tuple.
<code>
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:])
</code>
<br>
<b>Range of Negative Indexes</b>
<br>
• Specify negative indexes if you want to start the search from the end of the tuple.
<code>
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[-4:-1])
</code>
<br>
<b>Check if Item Exists</b>
<br>
• To determine if a specified item is present in a tuple use the in keyword.
<code>
thistuple = ("apple", "banana", "cherry")
if "apple" in thistuple:
  print("Yes, 'apple' is in the fruits tuple")
</code>
<br>
<b>Counting Tuple Items</b>
<br>
• The count() method returns the number of times a specified value appears in the tuple.
<code>
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
x = thistuple.count(5)
print(x)
</code>
<br>
<b>Indexing Tuple Items</b>
<br>
• The index() method finds the first occurrence of the specified value.
<br>
• The index() method raises an exception if the value is not found.
<code>
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
x = thistuple.index(8)
print(x)
</code>
<br>
<b>Update Tuples</b>
<br>
• Tuples are unchangeable meaning that you cannot change, add or remove items once the tuple is created.
<br>
• But there are some workarounds.
<br>
<br>
<b>Change Tuple Values</b>
<br>
• Once a tuple is created you cannot change its values. Tuples are unchangeable or immutable.
<br>
• But there is a workaround. You can convert the tuple into a list, change the list and convert the list back into a tuple.
<code>
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)
</code>
<br>
<b>Add Items</b>
<br>
• Since tuples are immutable, they do not have a built-in append() method but there are other ways to add items to a tuple.
<br>
<b>Convert into a list</b>
<br>
• Just like the workaround for changing a tuple you can convert it into a list, add your item(s) and convert it back into a tuple.
<code>
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
</code>
<b>Add tuple to a tuple</b>
<br>
• You are allowed to add tuples to tuples so if you want to add one or more items, create a new tuple with the items and add it to the existing tuple.<code>
thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y
print(thistuple)
</code>
<br>
<b>Remove Items</b>
<br>
• Tuples are unchangeable so you cannot remove items from it but you can use the same workaround as we used for changing and adding tuple items.
<code>
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
</code>
• Or you can delete the tuple completely.
<code>
thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple) #this will raise an error because the tuple no longer exists
</code>
<br>
<b>Unpack Tuples</b>
<br>
• When we create a tuple we normally assign values to it. This is called "packing" a tuple.
<code>
fruits = ("apple", "banana", "cherry")
print(fruits)
</code>
• We are also allowed to extract the values back into variables. This is called "unpacking" a tuple.
<code>
fruits = ("apple", "banana", "cherry")
(green, yellow, red) = fruits
print(green)
print(yellow)
print(red)
</code>
<b>Note :</b> The number of variables must match the number of values in the tuple, if not you must use an asterisk to collect the remaining values as a list.
<br>
• If the number of variables is less than the number of values, you can add an * to the variable name and the values will be assigned to the variable as a list.
<code>
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")
(green, yellow, *red) = fruits
print(green)
print(yellow)
print(red)
</code>
• If the asterisk is added to another variable name than the last, Python will assign values to the variable until the number of values left matches the number of variables left.
<code>
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")
(green, *tropic, red) = fruits
print(green)
print(tropic)
print(red)
</code>
<br>
<b>Loop Tuples</b>
<br>
• You can loop through the tuple items by using a for loop.
<code>
thistuple = ("apple", "banana", "cherry")
for x in thistuple:
  print(x)
</code>
<b>Loop Through the Index Numbers</b>
<br>
• You can also loop through the tuple items by referring to their index number.
<br>
• Use the range() and len() functions to create a suitable iterable.</b>
<code>
thistuple = ("apple", "banana", "cherry")
for i in range(len(thistuple)):
  print(thistuple[i])
</code>
<br>
<b>Join Tuples</b>
<br>
• To join two or more tuples you can use the + operator:
<code>
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)
tuple3 = tuple1 + tuple2
print(tuple3)
</code>
<b>Multiply Tuples</b>
<br>
• If you want to multiply the content of a tuple a given number of times, you can use the * operator.
<code>
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2
print(mytuple)
</code>
</p>

<h2 style="color:#6a66bd; text-align:left;">Python Sets</h2>

<p style="color:#6a66bd;">
• Sets are used to store multiple items in a single variable.
<br>
• Sets are written with curly brackets.
<br>
• A set is a collection which is unordered, unchangeable and unindexed.
<br>
<b>Note :</b> Set items are unchangeable, but you can remove items and add new items.
<code>
thisset = {"apple", "banana", "cherry"}
print(thisset)
</code>
<b>Note :</b> Sets are unordered, so you cannot be sure in which order the items will appear.
<br>
<br>
<b>Frozen Set</b>
<br>
• The frozenset() function returns an unchangeable frozenset object (which is like a set object, only unchangeable).
<code>
mylist = ['apple', 'banana', 'cherry']
x = frozenset(mylist)
</code>
• Try to change the value of a frozenset item. This will cause an error.
<code>
mylist = ['apple', 'banana', 'cherry']
x = frozenset(mylist)
x[1] = "strawberry" #frozenset object does not support item assignment
print(x)
</code>
<br>
<b>Set Items</b>
<br>
• Set items are unordered, unchangeable and do not allow duplicate values.
<br>
<br>
<b>Unordered</b>
<br>
• Unordered means that the items in a set do not have a defined order.
<br>
• Set items can appear in a different order every time you use them and cannot be referred to by index or key.
<br>
<br>
<b>Unchangeable</b>
<br>
• Set items are unchangeable meaning that we cannot change the items after the set has been created.
<br>
• Once a set is created you cannot change its items but you can remove items and add new items.</b>
<br>
<br>
<b>Duplicates Not Allowed</b>
<br>
• Sets cannot have two items with the same value.
<code>
thisset = {"apple", "banana", "cherry", "apple"} #Duplicates will be ignored in output
print(thisset)
</code>
<b>Note :</b> The values True and 1 are considered the same values in sets and are treated as duplicates.
<br>
• Similarly, False and 0 are considered the same values in sets and are treated as duplicates.
<br>
<br>
<b>Sets Length</b>
<br>
• To determine how many items a set has we use the len() function.
<code>
thisset = {"apple", "banana", "cherry"}
print(len(thisset))
</code>
<br>
<b>Set Items - Data Types</b>
<br>
• Set items can be of any data type.
<code>
set1 = {"apple", "banana", "cherry"}
set2 = {1, 5, 7, 9, 3}
set3 = {True, False, False}
</code>
• A set can contain different data types.
<code>
set1 = {"abc", 34, True, 40, "male"}
</code>
<br>
<b>set() Constructor</b>
<br>
• It is also possible to use the set() constructor to make a set.
<code>
thisset = set(("apple", "banana", "cherry")) #note the double round-brackets
print(thisset)
</code>
<br>
<b>Access Set Items</b>
<br>
• You cannot access items in a set by referring to an index or a key.
<br>
• But you can loop through the set items using a for loop or ask if a specified value is present in a set by using the in keyword.
<code>
thisset = {"apple", "banana", "cherry"}
for x in thisset:
  print(x)
</code>
<br>
<b>Add Set Items</b>
<br>
• To add one item to a set use the add() method.
<code>
thisset = {"apple", "banana", "cherry"}
thisset.add("orange")
print(thisset)
</code>
<b>Add Sets</b>
<br>
• To add items from another set into the current set use the update() method.
<code>
thisset = {"apple", "banana", "cherry"}
tropical = {"pineapple", "mango", "papaya"}
thisset.update(tropical)
print(thisset)
</code>
<b>Add Any Iterable</b>
<br>
• The object in the update() method does not have to be a set it can be any iterable object (tuples, lists, dictionaries etc.).
<code>
thisset = {"apple", "banana", "cherry"}
mylist = ["kiwi", "orange"]
thisset.update(mylist)
print(thisset)
</code>
<br>
<b>Remove Set Items</b>
<br>
• To remove an item in a set use the remove() or the discard() method.
<br>
• If the item to remove does not exist remove() will raise an error.
<code>
thisset = {"apple", "banana", "cherry"}
thisset.remove("banana")
print(thisset)
</code>
• If the item to remove does not exist, discard() will NOT raise an error.
<code>
thisset = {"apple", "banana", "cherry"}
thisset.discard("banana")
print(thisset)
</code>
• You can also use the pop() method to remove an item but this method will remove a random item so you cannot be sure what item that gets removed.
<br>
• The return value of the pop() method is the removed item.
<code>
thisset = {"apple", "banana", "cherry"}
x = thisset.pop()
print(x)
print(thisset)
</code>
• The clear() method empties the set.
<code>
thisset = {"apple", "banana", "cherry"}
thisset.clear()
print(thisset)
</code>
• The del keyword will delete the set completely.
<code>
thisset = {"apple", "banana", "cherry"}
del thisset
print(thisset)
</code>
<br>
<b>Loop Set Items</b>
<br>
• You can loop through the set items by using a for loop.
<code>
thisset = {"apple", "banana", "cherry"}
for x in thisset:
  print(x)
</code>
<br>
<b>Join Sets</b>
<br>
• There are several ways to join two or more sets in Python.
<br>
• The union() and update() methods joins all items from both sets.
<br>
• The intersection() method keeps ONLY the duplicates.
<br>
• The difference() method keeps the items from the first set that are not in the other set(s).
<br>
• The symmetric_difference() method keeps all items EXCEPT the duplicates.
<br>
<br>
<b>Union</b>
<br>
• The union() method returns a new set with all items from both sets.
<code>
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = set1.union(set2)
print(set3)
</code>
• You can use the | operator instead of the union() method and you will get the same result.
<code>
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = set1 | set2
print(set3)
</code>
<br>
<b>Join Multiple Sets</b>
<br>
• All the joining methods and operators can be used to join multiple sets.
<br>
• When using a method just add more sets in the parentheses separated by commas.
<code>
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}
myset = set1.union(set2, set3, set4)
print(myset)
</code>
• When using the | operator separate the sets with more | operators.
<code>
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}
myset = set1 | set2 | set3 |set4
print(myset)
</code>
<br>
<b>Join a Set and a Tuple</b>
<br>
• The union() method allows you to join a set with other data types like lists or tuples. The result will be a set.
<code>
x = {"a", "b", "c"}
y = (1, 2, 3)
z = x.union(y)
print(z)
</code>
<b>Note :</b> The  | operator only allows you to join sets with sets and not with other data types like you can with the union() method.
<br>
<br>
<b>Update</b>
<br>
• The update() method inserts all items from one set into another.
<br>
• The update() changes the original set and does not return a new set.
<code>
set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}
set1.update(set2)
print(set1)
</code>

print(set3)

set2 = {"google", "microsoft", "apple"}
set1.intersection_update(set2)

• The values True and 1 are considered the same value. The same goes for False and 0.

<br>
• The difference_update() method will also keep the items from the first set that are not in the other set but it will change the original set instead of returning a new set.
<code>
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set1.difference_update(set2)
print(set1)
</code>

<b>Symmetric Differences</b>
<br>

<h2 style="color:#6a66bd; text-align:left;">Python Dictionary</h2>

thisdict = {

  "model": "Mustang",
  "year": 1964
}
print(thisdict)
</code>

<b>Dictionary Items</b>
<br>
• Dictionary items are ordered, changeable and do not allow duplicates.

thisdict = {

<br>
<b>Duplicates Not Allowed</b>
<br>
• Dictionaries cannot have two items with the same key.
<code>
thisdict = {

}
print(thisdict)
</code>

<br>

thisdict = {
  "brand": "Ford",

  "year": 1964,

print(len(thisdict))

<br>

<b>Greater than :</b> <code>a > b</code>

<br>

a = 33

a = 33

if b > a:

elif a == b:
  print("a and b are equal")

<br>

</code>

<br>

<code>
a = 330

print("A") if a > b else print("=") if a == b else print("B")
</code>

<h2 style="color:#6a66bd; text-align:left;">Python Loops</h2>

<p style="color:#6a66bd;">
• Python has two primitive loops commands.
<br>
<b>While loop</b>
<br>
<b>For loop</b>
<br>
<br>
<b>While Loop</b>
<br>
• With the while loop we can execute a set of statements as long as a condition is true.

i = 1
while i < 6:
  print(i)
  i += 1
</code>
<b>Note :</b> Remember to increment i or else the loop will continue forever.
<br>
<b>Break Statement</b>
<br>
• With the break statement we can stop the loop even if the while condition is true.
<code>
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
</code>
<b>Continue Statement</b>
<br>
• With the continue statement we can stop the current iteration and continue with the next.
<code>
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
</code>
<br>
<b>For Loop</b>
<br>
• A for loop is used for iterating over a sequence (a list, a tuple, a dictionary, a set, or a string).
<br>
• With the for loop we can execute a set of statements once for each item in a list, tuple, set etc.
<code>
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
</code>
<b>Note :</b> The for loop does not require an indexing variable to set beforehand.
<br>
<b>Looping Through a String</b>
<br>
• Even strings are iterable objects they contain a sequence of characters.
<code>
for x in "banana":
  print(x)
</code>
<b>Break Statement</b>
<br>
• With the break statement we can stop the loop before it has looped through all the items.
<code>
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
</code>
<b>Continue Statement</b>
<br>
• With the continue statement we can stop the current iteration of the loop and continue with the next.
<code>
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
</code>
<br>
<b>Range() Function</b>
<br>
• To loop through a set of code a specified number of times we can use the range() function.
<br>
• The range() function returns a sequence of numbers starting from 0 by default and increments by 1 (by default) and ends at a specified number.
<code>
for x in range(6):
  print(x)
</code>
<b>Note :</b> The range(6) is not the values of (0 to 6) but the values (0 to 5).
<br>
• The range() function defaults to 0 as a starting value however it is possible to specify the starting value by adding a parameter : <code>range(2, 6)</code>.
<br>
• The range() function defaults to increment the sequence by 1 however it is possible to specify the increment value by adding a third parameter : <code>range(2, 30, 3)</code>.
<br>
<br>
<b>Nested Loops</b>
<br>
• A nested loop is a loop inside a loop.
<br>
• The "inner loop" will be executed one time for each iteration of the "outer loop".
<code>
adj = ["red", "big", "tasty"]

<h2 style="color:#6a66bd; text-align:left;">Python Functions</h2>

<p style="color:#6a66bd;">
• A function is a block of code which only runs when it is called.
<br>
• You can pass data known as parameters into a function.
<br>
• A function can return data as a result.
<br>
<b>Creating a Function</b>
<br>
• In Python a function is defined using the def keyword.
<code>
def my_function():
  print("Hello from a function")
</code>
<b>Calling a Function</b>
<br>
• To call a function use the function name followed by parenthesis.
<code>
def my_function():
  print("Hello from a function")
my_function()
</code>
<br>
<b>Arguments</b>
<br>
• Information can be passed into functions as arguments.
<br>
• Arguments are specified after the function name inside the parentheses. You can add as many arguments as you want just separate them with a comma.
<code>
def my_function(fname):
  print(fname + " Refsnes")
my_function("Emil")
my_function("Tobias")
my_function("Linus")
</code>
<b>Note :</b> Arguments are often shortened to args in Python documentations.
<br>
• A function must be called with the correct number of arguments.
<br>
• If your function expects 2 arguments you have to call the function with 2 arguments not more and not less. 
<br>
<b>Arbitrary Arguments</b>
<br>
• If you do not know how many arguments that will be passed into your function add a * before the parameter name in the function definition.
<br>
• This way the function will receive a tuple of arguments and can access the items accordingly.
<code>
def my_function(*kids):
  print("The youngest child is " + kids[2])
my_function("Emil", "Tobias", "Linus")
</code>
<b>Keyword Arguments</b>
<br>
• You can also send arguments with the key = value syntax.
<br>
• This way the order of the arguments does not matter.
<code>
def my_function(child3, child2, child1):
  print("The youngest child is " + child3)
my_function(child1 = "Emil", child2 = "Tobias", child3 = "Linus")
</code>
<b>Default Parameter Value</b>
<br>
• If we call the function without argument it uses the default value:
<code>
def my_function(country = "Norway"):
  print("I am from " + country)
my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
</code>
<b>Difference between Parameter and Arguments</b>
<br>
• A Parameter is the variable listed inside the parentheses in the function definition.
<br>
• An Argument is the value that is sent to the function when it is called.
<br>
<b>Return Values</b>
<br>
• To let a function return a value use the return statement.
<code>
def my_function(x):

print(my_function(9))
</code>
<br>
<b>Pass Statement</b>

def myfunction():
  pass
</code>
<br>
<b>Positional-Only Arguments</b>
<br>
• You can specify that a function can have ONLY positional arguments or ONLY keyword arguments.
<br>

<code>
def my_function(x, /):
  print(x)
my_function(3)
</code>

<code>
def my_function(x):
  print(x)
my_function(x = 3)
</code>
<br>
<b>Keyword-Only Arguments</b>
<br>
• To specify that a function can have only keyword arguments add * before the arguments.
<code>
def my_function(*, x):
  print(x)
my_function(x = 3)
</code>
• Without the * you are allowed to use positional arguments even if the function expects keyword arguments.
<code>
def my_function(x):
  print(x)
my_function(3)
</code>

<b>Recursion</b>
<br>
• Python also accepts function recursion which means a defined function can call itself.
<br>
• Recursion is a common mathematical and programming concept. 
<br>
• It means that a function calls itself. This has the benefit of meaning that you can loop through data to reach a result.
<code>
def tri_recursion(k):
  if(k > 0):
    result = k + tri_recursion(k - 1)

  return result
print("Recursion Example Results")
tri_recursion(6)

<h2 style="color:#6a66bd; text-align:left;">Python Scope</h2>

<p style="color:#6a66bd;">
• A variable is only available from inside the region it is created. This is called scope.
<br>
<br>
<b>Local Scope</b>
<br>
• A variable created inside a function belongs to the local scope of that function and can only be used inside that function.
<code>
def myfunc():
  x = 300
  print(x)
myfunc()
</code>
<br>
<b>Global Scope</b>
<br>
• A variable created in the main body of the Python code is a global variable and belongs to the global scope.
<br>
• Global variables are available from within any scope global and local.
<code>
x = 300
def myfunc():
  print(x)
myfunc()
print(x)
</code>
<br>
<b>Naming Variable</b>
<br>
• If you operate with the same variable name inside and outside of a function Python will treat them as two separate variables one available in the global scope (outside the function) and one available in the local scope (inside the function).
<code>
x = 300
def myfunc():
  x = 200
  print(x)
myfunc()
print(x)
</code>
<br>
<b>Global Keyword</b>
<br>

<h2 style="color:#6a66bd; text-align:left;">Python Try Except</h2>

<p style="color:#6a66bd;">
• The try block lets you test a block of code for errors.
<br>
• The except block lets you handle the error.
<br>
• The else block lets you execute code when there is no error.
<br>
• The finally block lets you execute code regardless of the result of the try- and except blocks.
<br>
<br>
<b>Exception Handling</b>
<br>
• When an error occurs or exception as we call it, Python will normally stop and generate an error message.
<br>
These exceptions can be handled using the try statement.
<code>
try:
  print(x)
except:
  print("An exception occurred")
</code>
<b>Note :</b> Since the try block raises an error the except block will be executed.
<br>
• Without the try block the program will crash and raise an error.
<br>
<br>
<b>Many Exceptions</b>
<br>
• You can define as many exception blocks as you want e.g. if you want to execute a special block of code for a special kind of error.
<br>
• Print one message if the try block raises a NameError and another for other errors.
<code>
try:
  print(x)
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
</code>
<b>Else</b>
<br>
• You can use the else keyword to define a block of code to be executed if no errors were raised.
<code>
try:
  print("Hello")
except:
  print("Something went wrong")
else:
  print("Nothing went wrong")
</code>
<b>Finally</b>
<br>
• The finally block if specified will be executed regardless if the try block raises an error or not.
<code>
try:
  print(x)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
</code>
<b>Raise an Exception</b>
<br>
• As a Python developer you can choose to throw an exception if a condition occurs.
<br>
• To throw (or raise) an exception use the raise keyword.
<code>
x = -1
if x < 0:
  raise Exception("Sorry, no numbers below zero")
</code>
• You can define what kind of error to raise and the text to print to the user.
<code>
x = "hello"
if not type(x) is int:
  raise TypeError("Only integers are allowed")
</code>
</p>

<h2 style="color:#6a66bd; text-align:left;">Python File Handling</h2>

<p style="color:#6a66bd;">
• Python has several functions for creating, reading, updating and deleting files.
<br>
• The key function for working with files in Python is the open() function.
<br>
• The open() function takes two parameters : filename and mode.
<br>
• There are four different methods for opening a file :
<br>
<b>Read (r) </b>- Default value. Opens a file for reading, error if the file does not exist.
<br>
<b>Append (a) </b>- Opens a file for appending, creates the file if it does not exist.
<br>
<b>Write (w) </b>- Opens a file for writing, creates the file if it does not exist
<br>
<b>Create (x) </b>- Creates the specified file, returns an error if the file exists.
<br>
• In addition you can specify if the file should be handled as binary or text mode :
<br>
<b>Text (t)</b> - Default value. Text mode.
<br>
<b>Binary (b)</b> - Binary mode.
<br>
<br>
<b>Opening a File</b>
<br>
• To open a file for reading it is enough to specify the name of the file.
<code>
f = open("demofile.txt")
</code>
But you can also do that.
<code>
f = open("demofile.txt", "rt")
</code>
<b>Note :</b> Here "r" for read and "t" for text are the default values, you do not need to specify them.
<br>
• Make sure the file exists or else you will get an error.
<br>
<br>
<b>Opening a File on the Server</b>
<br>
• Assume we have the following file <code>demofile.txt</code> located in the same folder as Python.
<br>
• To open the file use the built-in open() function.
<br>
• The open() function returns a file object which has a read() method for reading the content of the file.
<code>
f = open("demofile.txt", "r")
print(f.read())
</code>
• If the file is located in a different location you will have to specify the file path.
<code>
f = open("D:\\myfiles\demofile.txt", "r")
print(f.read())
</code>
<br>
<b>Read Only Parts of the File</b>
<br>
• By default the read() method returns the whole text but you can also specify how many characters you want to return.
<code>
f = open("demofile.txt", "r")
print(f.read(5))
</code>
<br>
<b>Read Lines</b>
<br>
• You can return one line by using the readline() method.
<code>
f = open("demofile.txt", "r")
print(f.readline())
</code>
• By calling readline() two times you can read the two first lines.
<code>
f = open("demofile.txt", "r")
print(f.readline())
print(f.readline())
</code>
<br>
<b>Close Files</b>
<br>
• It is a good practice to always close the file when you are done with it.
<code>
f = open("demofile.txt", "r")
print(f.readline())
f.close()
</code>
<b>Note :</b> You should always close your files, in some cases due to buffering changes made to a file may not show until you close the file.
<br>
<br>
<b>Write a File</b>
<br>
• To write to an existing file you must add a parameter to the open() function.
<br>
<b>Append (a)</b> - will append to the end of the file
<br>
<b>Write (w)</b> - will overwrite any existing content
<code>
f = open("demofile2.txt", "a")
f.write("Now the file has more content!")
f.close()
#open and read the file after the appending
f = open("demofile2.txt", "r")
print(f.read())
</code>
<br>
<b>Create a New File</b>
<br>
• To create a new file in Python use the open() method with one of the following parameters :
<br>
<b>Create (x)</b> - will create a file, returns an error if the file exist
<br>
<b>Append (a)</b> - will create a file if the specified file does not exist
<br>
<b>Write (w)</b> - will create a file if the specified file does not exist
<code>
f = open("myfile.txt", "x")
</code>
<br>
<b>Delete a File</b>
<br>
• To delete a file you must import the OS module and run its os.remove() function.
<code>
import os
os.remove("demofile.txt")
</code>
<b>Note : </b>To delete an entire folder, use the os.rmdir() method.
<code>
import os
os.rmdir("myfolder")
</code>
</p>
